```python
import pandas as pd

# 1. Leggi il CSV
csv_input = "audio_representation_wav2vec.csv"   # Nome del CSV di input
df = pd.read_csv(csv_input)

# 2. Aggiungi la colonna 'snr' = 0 (per tutte le righe)
df['snr'] = 0

# 3. Salva in un nuovo CSV
csv_output = "audio_representation_wav2vec.csv"
df.to_csv(csv_output, index=False)

print(f"Creato il file {csv_output} con colonna 'snr' impostata a 0.")

```
